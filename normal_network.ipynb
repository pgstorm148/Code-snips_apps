```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Define the number of normal data points you want
num_normal_samples = 10000

# Create a DataFrame to store the data
data = pd.DataFrame(columns=['source_ip', 'dest_ip', 'source_port', 'dest_port', 'protocol', 'timestamp', 'bytes_transferred', 'packets_sent', 'label'])

# Generate random normal network traffic data
for _ in range(num_normal_samples):
    source_ip = f"192.168.{random.randint(1, 254)}.{random.randint(1, 254)}"
    dest_ip = f"10.0.{random.randint(1, 254)}.{random.randint(1, 254)}"
    source_port = random.randint(1024, 65535)
    dest_port = random.randint(1, 1023)
    protocol = random.choice(['TCP', 'UDP', 'ICMP'])
    timestamp = datetime.now() - timedelta(seconds=random.randint(1, 3600))
    bytes_transferred = random.randint(1, 10000)
    packets_sent = random.randint(1, 100)
    label = 0  # 0 for normal

    data = data.append({'source_ip': source_ip,
                        'dest_ip': dest_ip,
                        'source_port': source_port,
                        'dest_port': dest_port,
                        'protocol': protocol,
                        'timestamp': timestamp,
                        'bytes_transferred': bytes_transferred,
                        'packets_sent': packets_sent,
                        'label': label}, ignore_index=True)

# Save the normal data to a CSV file
data.to_csv('normal_traffic.csv', index=False)
```

```python
# Create anomalies by modifying some normal data
num_anomalous_samples = 500

for _ in range(num_anomalous_samples):
    # Select a random normal data point and modify it to create an anomaly
    idx = random.randint(0, len(data) - 1)
    data.at[idx, 'bytes_transferred'] = random.randint(10001, 50000)
    data.at[idx, 'label'] = 1  # 1 for anomaly

# Save the dataset with anomalies to a CSV file
data.to_csv('network_traffic_anomalies.csv', index=False)
```
